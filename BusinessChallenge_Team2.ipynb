```python
# import the dataset 

import pandas as pd #importing pandas
import numpy as np # importing numpy
import matplotlib.pyplot as plt # importing matlab visualization library
import seaborn as sns # importing seaborn



file= './Price.xlsx' # importing the file
Price_df= pd.read_excel(file) #creating first df

# showing first rows of worksheet_df

Price_datatype= {'Total_cost': float,
                 'Price_Per_QTY': float, 
                 'Quantity': int, 
                 'Department': str,
                 'Vendor': str, 
                 'Object': str
                } 
Price_df = Price_df.astype(Price_datatype)

Price_df.head(n=5)
```

```python
def format_currency(value):
    """Formats the numerical Values into a currency."""
    return "${:,.2f}".format(value)


def Wedding_Cost_Breakdown(Prize_df): 
    
    #Department Level price summary
    department_summary = Price_df.groupby('Department').sum()
    
    #Vendor Level Summary 
    vendor_summary= Prize_df.groupby('Vendor').sum()
    
    
    tot_price = """
Client Wedding Proposal

"""
    
    
    #Itemized Breakdown 
    tot_price += """
Detailed Itemized Bill

"""
    for department,department_items in Price_df.groupby('Department'): 
        tot_price += f"\n Department: {department}\n"
        subtotal_department=0
        for index, row in department_items.iterrows():
            item = f" {row ['Vendor']} - {row ['Object']} (x {row['Quantity']})"
            tot_price+= f" {item}: {format_currency(row['Total_cost'])}\n"
            subtotal_department += row['Total_cost']
        tot_price += f" Subtotal for {department}: {format_currency(subtotal_department)}\n"
        
    #Total Price
    total_price = Price_df.Total_cost.sum()
    
    tot_price+= f"""

The total for the wedding is {format_currency (total_price)}

Please sign below to accept proposal: 


______________________________________"""
    

    return tot_price 

Wedding_cost_summary = Wedding_Cost_Breakdown(Price_df)
print(Wedding_cost_summary)
```
